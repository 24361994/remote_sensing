```python
plants = cv2.imread('/content/drive/My Drive/NDVI/04_rgb.jpeg')
```

```python
hsv = cv2.cvtColor(plants, cv2.COLOR_RGB2HSV)
```

```python
plt.imshow(hsv)
plt.show()
```

```python
lower_green = np.array([75,110,80])
upper_green = np.array([100,255,255])
```

# Arreglar el umbral de color para la segmentaci√≥n

```python
mask = cv2.inRange(hsv, lower_green, upper_green)
```

* Para la imagen hsv, analizar y buscar todos los colores que e ajusten a mi umbral establecido y establece en base a eso una mascara

* Para mi imagen original, aplicale la mascara generada y elimina lo que no esta dentro de ella.

```python
result = cv2.bitwise_and(plants, plants, mask= mask)
```

* Muestra de la aplicaci√≥n de la mascara y transformaci√≥n de espacio BGR (por defecto) hacia RGB (formato original)

```python
hsv2 = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
plt.imshow(hsv2)
plt.show()
```
